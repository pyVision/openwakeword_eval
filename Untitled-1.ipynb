```python

import torch
import torchaudio
import matplotlib.pyplot as plt

def plot_waveform(waveform, title="Waveform"):
    plt.figure(figsize=(10, 4))
    plt.plot(waveform.numpy())
    plt.title(title)
    plt.show()

def get_audio_duration(file_path):
    metadata = torchaudio.info(file_path)
    duration = metadata.num_frames / metadata.sample_rate
    return duration

def plot_reverb(audio,reverb,title1,title2):

    audio_tensor, sample_rate1 = torchaudio.load(audio)
    enhanced_reverb_audio, sample_rate2 = torchaudio.load(reverb)

    duration = get_audio_duration(audio)
    print(f"Duration: {duration} seconds")

    duration = get_audio_duration(reverb)
    print(f"Duration: {duration} seconds")

    print("srate",sample_rate1,sample_rate2)

    # Plot original waveform
    plot_waveform(audio_tensor.squeeze(0), title=title1)

    # Plot reverberated waveform
    plot_waveform(enhanced_reverb_audio.squeeze(0)[0:16000], title=title2)


plot_reverb("aout1/0bb3d1f962b144b2928a5a5571332a5c.wav","S11/0bb3d1f962b144b2928a5a5571332a5c_1.2.wav","original","augmented")
```
